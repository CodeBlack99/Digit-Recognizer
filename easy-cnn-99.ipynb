```python
sample_sub = pd.read_csv("/kaggle/input/digit-recognizer/sample_submission.csv")
train_data = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
test_data = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
```

```python
x_train = train_data.drop(["label"], axis = 1)
y_train = train_data["label"]
```

```python
x_test = test_data
y_test = sample_sub
```

```python
missing_vals = x_train.isnull().sum()
print(missing_vals[missing_vals > 0])
```

```python
missing_vals = x_test.isnull().sum()
print(missing_vals[missing_vals > 0])
```

```python
x_test.shape
```

```python
y_train = pd.get_dummies(y_train).values
```

```python
from sklearn.model_selection import train_test_split

x, x_val, y, y_val = train_test_split(x_train, y_train, test_size = 0.1, random_state=42)
```

```python
x.shape
```

```python
y_val.shape
```

```python
y.shape
```

```python
from keras import layers
from keras import models
from keras.datasets import mnist
from keras.utils import to_categorical

# Define the CNN model
model = models.Sequential()

model.add(layers.Conv2D(32, (5,5), activation='relu',padding = 'Same', input_shape=(28, 28,1)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (5, 5), padding = 'Same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(256, (5, 5), padding = 'Same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (5, 5), padding = 'Same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (5, 5), padding = 'Same', activation='relu'))
model.add(layers.Flatten())
model.add(Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))

model.summary()
```

```python
x = np.array(x).reshape(x.shape[0], 28, 28, 1)
```

```python
x.shape
```

```python
train_images = x
train_images = train_images.astype('float32') / 255
```

```python
x_test = np.array(x_test).reshape(x_test.shape[0], 28, 28, 1)
```

```python
test_images = x_test
test_images = test_images.astype('float32') / 255

test_images.shape
```

```python
x_val = np.array(x_val).reshape(x_val.shape[0], 28, 28, 1)
```

```python
x_val = x_val.astype('float32') / 255

x_val.shape
```

```python
y.shape
```

```python
y_train.shape
```

```python
y
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
# Use the training data to train the model
model.fit(train_images, y,
          validation_data = (x_val,y_val),
          batch_size=100,
          epochs=20,
          verbose=1)

```

```python
preds_test = model.predict(x_test)
```

```python
preds_test.shape
```

```python
preds_test = np.argmax(preds_test,axis = 1)

results = pd.Series(preds_test,name="Label")
```

```python
output = pd.DataFrame({'ImageId': test_id,
                       'Label': preds_test})

output.to_csv('submission.csv', index=False)
```
